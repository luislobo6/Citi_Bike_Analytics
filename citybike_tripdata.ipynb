```python
import pandas as pd
```

```python
from os import listdir

files = listdir("data/")
```

```python
dataframes = {}

keys = []

for file in files:
    key = file[0:9]
    keys.append(key)
    
    dataframes[key] = pd.read_csv(f"data/{file}")
```

```python
concatenate = []

for key in keys:
    concatenate.append(dataframes[key])
```

```python
citybike_tripdata = pd.concat(concatenate, axis=0, ignore_index=True)
```
